```python
import pandas as pd
import matplotlib.pyplot as plt

# Manually input the data you shared
data = {
    "K": [2, 3, 4, 5, 6, 7, 8, 9, 10],
    "WSSSE": [
        1748529.602, 1452806.735, 1317582.592, 1203555.979, 1097472.702,
        1019806.918, 1005972.610, 894631.8437, 848847.6728
    ]
}

df_elbow = pd.DataFrame(data)

# Plot the Scree Plot
plt.figure(figsize=(8, 5))
plt.plot(df_elbow["K"], df_elbow["WSSSE"], marker='o', linestyle='-')
plt.title("Scree Plot (Elbow Method) for KMeans Clustering")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Within-Cluster Sum of Squared Errors (WSSSE)")
plt.xticks(df_elbow["K"])
plt.grid(True)
plt.tight_layout()
plt.show()
```
